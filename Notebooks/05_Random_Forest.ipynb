```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
gr_train = pd.read_csv('/content/drive/MyDrive/training_set_finale.csv')
gr_test= pd.read_csv('/content/drive/MyDrive/test_df.csv')
```

```python
#this fuction extract only words in lemmatize text without considering others special characters.

import re
compiler = re.compile("\w+")

def fixing_column(df, column):
    global compiler
    return [" ".join(compiler.findall(row)) for row in df[column]]
```

```python
gr_train.lemmatized_text = fixing_column(gr_train, "lemmatized_text")
gr_test.lemmatized_text = fixing_column(gr_test, "lemmatized_text")

print(gr_train.lemmatized_text[0], "\n")
print(gr_test.lemmatized_text[0])
```

```python
gr_test.head(3)
```

```python
X_train = gr_train.lemmatized_text.values
X_test = gr_test.lemmatized_text.values

#saving the target variable
label_train= gr_train['genre']
label_test= gr_test['genre']

#the model will be trained only on the 'lemmatized_text' column
col_to_del= ['genre','book_id','rating', 'review_text']
gr_lemmatized= gr_train.drop(columns= col_to_del) #in case you want to execute train/test split per validation set
```

```python
X_train[0]
```

```python
# split the training set in training and validation set. The model is tested on the validation set
# X_train, X_test, y_train y_test = train_test_split(gr_lemmatized, label_train, test_size=0.30, random_state=42)
```

```python
# implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(1,1))
traindataset=countvector.fit_transform(X_train)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='gini')
randomclassifier.fit(traindataset, label_train)
```

```python
# Predict for the Test_set

test_datasetCV = countvector.transform(X_test) #.lemmatized_text))
predictionsCV = randomclassifier.predict(test_datasetCV)
```

```python
matrix=confusion_matrix(label_test,predictionsCV)
print(matrix, "\n")
score=accuracy_score(label_test,predictionsCV)
print(score, "\n")
report=classification_report(label_test,predictionsCV)
print(report)
```

## TF-IDF

TF-IDF measures the relative importance of words by comparing the number of times a word appear in a sample with the number of samples with that word.

```python
# implement BAG OF WORDS
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
traindatasetTF=tfidfvector.fit_transform(X_train)#.lemmatized_text))
```
